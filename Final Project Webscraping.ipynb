```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_the_busiest_airports_in_the_United_States"
```

```python
scrapage = requests.get(url)
scrapage.status_code
```

```python
soup = BeautifulSoup(scrapage.content, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
soup.select('table:nth-child(9)')
```

```python
print(soup.select('table:nth-child(9)'))
```

```python
soup.select('table:nth-child(9)')[0].get_text()
```

```python
table = soup.find('table',class_='wikitable sortable')
table
```

```python
airports = pd.read_html(str(table))[0]
airports
```

```python
airports = pd.DataFrame(airports)
airports.head()
```

```python
new_columns = {'Rank (2022)':'Rank','Airports (large)':'Airport','IATA Code':'Code','Major cities served':'City','Metro Area':'Metro','2022[2]':'2022','2021[3]':'2021','2020[4]':'2020','2019[5]':'2019','2018[6]':'2018','2017[7]':'2017','2016[8]':'2016','2015[9]':'2015','2014[10]':'2014','2013[11]':'2013'}
airports.rename(columns=new_columns, inplace=True)
```

```python
airports
```

```python
airports.to_csv('busiest_airports2022.csv', index=False)
```
